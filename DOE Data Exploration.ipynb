```python
import pandas as pd
import csv

df = pd.read_csv('C:\\Users\\SarahKhederian\\Documents\\DoE Project\\statefundingassessment.csv', sep=",", engine = 'python')


df.head(10)

```

Derive New Variables

```python
#Encode State Labels
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
label_state = pd.DataFrame(le.fit_transform(df.State))
state = pd.concat([df.State, label_state], axis=1)

state.head()

df2 = pd.concat([df, label_state], axis=1)
del df2['State']
df2.head(50)
```

```python
#Calculate all funding per capita

df2.dtypes

for column in df2:
    df2['%s_percap' % column] = df2[column]/df2['State Population']
```

```python
#Comparing Funding to 4th Grade Assessment Scores

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.pairplot(df2, x_vars=['Total_percap','State Total_percap', 'Federal Special education_percap', 'state special education_percap'], y_vars='4th Grade Scale Score (2015)', size=7, aspect=.7)
```
